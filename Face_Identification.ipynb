```python
import cv2
import os

# Download the XML file if it doesn't exist
xml_file_path = '/content/haarcascade_frontalface_default.xml'
if not os.path.exists(xml_file_path):
  !wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml -O $xml_file_path
  print(f"Downloaded haarcascade_frontalface_default.xml to {xml_file_path}")

# Load the face cascade
face_cascade = cv2.CascadeClassifier(xml_file_path)

# Continue with your code...
img = cv2.imread('/content/mahesh.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.05, minNeighbors=5)
```

```python
print(faces)
```

```python
for x,y,w,h in faces:
  img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 3)
```

```python
cv2_imshow(img)
```

```python
img = cv2.imread('/content/couple.jpg')
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
edged = cv2.Canny(gray_img, 30, 200)
```

```python
contours , hierarchy = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img, contours, -1, (0,255,0), 2)
```
